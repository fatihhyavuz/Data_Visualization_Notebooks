```python
import numpy as np
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns 
%matplotlib inline
import warnings;
warnings.filterwarnings("ignore")
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe().T
```

## scatter plot

```python
sns.scatterplot(x= "total_bill", y="tip", data=tips )
plt.show()
```

```python
# regresyon cizgisi cevresindeki golge ne kadar darsa  tahminleri o kadar yanilma payi dusuk olur 
sns.regplot(data=tips, y= "tip" ,x="total_bill") 
plt.show()
```

```python
sns.scatterplot(x= "total_bill", y="tip", data=tips, hue = "sex" )
plt.show()
```

```python
sns.scatterplot(x= "total_bill", y="tip", data=tips, hue = "smoker" )
plt.show()
```

```python
sns.scatterplot(x= "total_bill", y="tip", data=tips, hue = "day" )
plt.show()
```

## style

```python
sns.scatterplot(x= "total_bill", y="tip", data=tips, hue = "smoker",style= "sex" )
plt.show()
```

```python
sns.scatterplot(x= "total_bill", y="tip", data=tips, hue = "smoker",style= "day" )
plt.show()
```

```python
sns.scatterplot(x= tips["total_bill"], y=tips["tip"], hue = tips["smoker"],style= tips["sex"] )
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize = (10,5))
sns.scatterplot(x="total_bill",y="tip",data=tips, style ="sex" , ax= ax[0])
sns.scatterplot(x="total_bill",y="tip",data=tips, hue ="smoker" , ax= ax[1]);
```

```python
# palette    https://seaborn.pydata.org/tutorial/objects_interface.html
plt.figure(figsize= (10,5))

sns.scatterplot(x="total_bill" ,y="tip",data =tips, hue ="sex" ,palette = "plasma");
```

```python
plt.figure(figsize= (10,5))

sns.scatterplot(x="total_bill" ,y="tip",data =tips, hue ="sex" ,palette = "plasma", alpha= 0.3);
```

```python
plt.figure(figsize= (10,5))

sns.scatterplot(x="total_bill" ,
                y="tip",data =tips,
                hue ="sex" ,
                palette = "plasma",
                alpha= 0.7,
               s = 300);
```

```python
import numpy as np
import pandas as pd 
import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns 
%matplotlib inline
import warnings;
warnings.filterwarnings("ignore")
```

# Distibution Plot

```python
tips = sns.load_dataset("tips")
```

```python
sns.displot(x="total_bill",data= tips)

plt.show()
```

```python
sns.displot(x="total_bill",data= tips, bins = 50) #  dagilim daha net gorulur oluyor 

plt.show()
```

```python
sns.displot(x="total_bill",data= tips, bins = 50, kde = True) 

plt.show()
```

```python
fig , ax = plt.subplots(1,2,figsize = (10,5))

sns.scatterplot(x="total_bill", y = "tip", data = tips, s=200, hue= "sex", ax= ax[0])
sns.histplot(x="total_bill", data = tips, kde = True, ax = ax[1]);
```

```python
sns.displot(x = "total_bill", data = tips, kde = True, kind='hist', aspect= 5) # aspect grafigi yayar
plt.show()
```

# histplot

```python
plt.figure(figsize = (5, 5))
sns.histplot(x = "total_bill", data = tips);
```

```python
plt.figure(figsize = (5, 5))
sns.histplot(x = "total_bill", data = tips, kde = True);
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 5))
sns.rugplot(x = "total_bill", data = tips, ax = ax[0])
sns.histplot(x = "total_bill", data = tips, kde = True, ax = ax[1]);
```

```python
sns.set(style = "dark")
sns.histplot(x = "total_bill", data = tips, kde = True);
```

```python
fig, ax = plt.subplots(1, 2, figsize = (10, 5))
sns.rugplot(x = "total_bill", data = tips, ax = ax[0])
sns.histplot(x = "total_bill", data = tips, kde = True, ax = ax[1]);
```

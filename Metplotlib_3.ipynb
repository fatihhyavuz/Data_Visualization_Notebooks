# gunluk kazanc bulma (feature engineering)

```python
tips.groupby("day").sum()
```

```python
numeric_tips = tips.select_dtypes(include = ["float64", "int64"])
numeric_tips.groupby(tips["day"]).sum()
```

```python
day = numeric_tips.groupby(tips["day"])["total_bill"].sum().index
```

```python
# Sayısal verilerin seçilmesi
numeric_tips = tips[["total_bill", "day"]]
# Günler bazında "total_bill" toplamlarının seçilmesi
day_of_total_bill_sum = numeric_tips.groupby("day")["total_bill"].sum()
# Bar Grafiği için günler ve toplamlarının alınması 
days = day_of_total_bill_sum.index
total_bills = day_of_total_bill_sum.values
# Bar Grafiğin çizilmesi
fig, ax = plt.subplots(figsize = (8,6))
ax.bar(days, total_bills);
```

```python
# Bar Grafiğin çizilmesi
fig, ax = plt.subplots(figsize = (8,6))
ax.bar(days, total_bills)
ax.set_title("Total Bill by Day")
ax.set_xlabel("Day")
ax.set_ylabel("Total Bill")
plt.show()
```

## pie chart

```python
x = np.array([100, 100, 100, 100])
mylabels= ("Apples", "Bananas", "Cherries", "Melon")
plt.figure(figsize = (10, 8))
plt.pie(x, labels = mylabels, labeldistance = 0.5, autopct = "%.2f")
plt.show()
```

```python
# explode parametresi dilimi ayirr autopct virgulden sonra kac basamak belirler, labeldistance labelin merkezden ne kadar uzak olacagini belirler
# startangle ile dondurulebilir 
x = np.array([100, 100, 100, 100])
mylabels= ("Apples", "Bananas", "Cherries", "Melon")
myexplode = [0.8,0,0.3,0]
plt.figure(figsize = (10, 8))
plt.pie(x, labels = mylabels, labeldistance = 0.5, autopct = "%.2f", explode = myexplode)
plt.show()
```

```python
x = np.array([100, 100, 100, 100])
mylabels= ("Apples", "Bananas", "Cherries", "Melon")
myexplode = [0.8,0,0.3,0]
plt.figure(figsize = (10, 8))
plt.pie(x, labels = mylabels, labeldistance = 0.5, autopct = "%.2f", startangle =90 )
plt.show()
```

```python

x = np.array([100, 100, 100, 100])
color_list = ["blue","red","pink","green"]
mylabels= ("Apples", "Bananas", "Cherries", "Melon")
myexplode = [0.8,0,0.3,0]
plt.figure(figsize = (10, 8))
plt.pie(x, 
        labels = mylabels,
        labeldistance = 0.5,
        autopct = "%.2f",
        startangle =90,
        colors = color_list)
plt.show()
```

```python
plt.figure(figsize = (10, 8))
color_list = ["gold", "yellowgreen", "lightcoral", "lightskyblue"]
explode_list = [0.5, 0.0, 0.0, 0.0]
y = np.array([25, 35, 25, 15])
my_labels = ("Apples", "Bananas", "Cherries", "Melon")
plt.pie(y, 
        labels = my_labels, 
        labeldistance = 0.5, 
        autopct = "%.1f", 
        startangle = 90, 
        pctdistance = 0.3, 
        colors = color_list, 
        explode = explode_list)
plt.legend(loc = (1))
plt.show()
```

## Datadan grafik olusturma 

```python
plt.figure(figsize=(10, 8))
sex_counts = tips["sex"].value_counts()
color_list = ["lightskyblue", "lightcoral"]
explode_list = [0.1, 0]
plt.pie(sex_counts,
        labels = sex_counts.index,
        labeldistance = 0.3,
        autopct = "%.1f",
        startangle = 140,
        shadow = True,
       pctdistance = 0.5,
       colors = color_list,
       explode = explode_list)
plt.legend(loc = (1))
plt.show()
```
